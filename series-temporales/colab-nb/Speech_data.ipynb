# Datos de habla.

```r
suppressPackageStartupMessages({
  library(astsa)
  library(tsibble)
  library(ggplot2)
  library(fable)
  library(feasts)
})
```

La serie consiste de una muestra de 1000 puntos tomados en el intervalo de $0{,}1$ segundos de habla registrada para la frase $aaa \ldots hhh$, donde se observa un patron de repeticion  de la señal y preiodicidades regulares. En el area del reconocimeinto del habla, se requiere poder realizar un analisis especteral para producir una huella de esta frase que puede luego compararse con huellas de otras silabas en una biblioteca de señales para buscar un _match_ o coincidencia. 

```r
speech |> 
  as_tsibble() |>
  autoplot(value) +
    ggtitle("Datos de Habla") +
    xlab("Segundos") +
    ylab("") +
    theme_light()
```

La separacion de las señales se conoce como periodo _pitch_ y representa la respuesta del filtro de tracto vocal a una secuencia periodica de pulsos simulados por la apertura y cerrado de la glotis.
